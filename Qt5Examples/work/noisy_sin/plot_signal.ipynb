```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
t=np.array(range(0,256))
```

```python
len(t)
```

```python
a=3
per=40
```

```python
noise=np.random.normal(0, 1, len(t))
```

```python
len(noise)
```

```python
s=a*np.sin(2*np.pi/per*t)+noise
```

```python
s_pure=a*np.sin(2*np.pi/per*t)
```

```python
plt.plot(t, s)
plt.plot(t, s_pure)
```
